```python
from pycaret.datasets import get_data
from pycaret.classification import *
import pandas as pd

# Load sample dataset
data = get_data('iris')

# Setup environment
s = setup(data, target='species', session_id=123, use_gpu=False)

# Train and get the best model automatically
best = compare_models()

# Save the model
save_model(best, 'iris_pycaret_model')
```

```python
import gradio as gr
from pycaret.classification import load_model, predict_model

# Load model for inference
model = load_model('iris_pycaret_model')

# Define prediction function
def predict_iris(sepal_length, sepal_width, petal_length, petal_width):
    sample = pd.DataFrame([[sepal_length, sepal_width, petal_length, petal_width]],
                          columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    prediction = predict_model(model, data=sample)
    label = prediction.loc[0, 'prediction_label']
    return f"ðŸŒ¼ Predicted Species: **{label}**"

# Build Gradio app
demo = gr.Interface(
    fn=predict_iris,
    inputs=[
        gr.Slider(4.0, 8.0, value=5.1, label="Sepal Length (cm)"),
        gr.Slider(2.0, 4.5, value=3.5, label="Sepal Width (cm)"),
        gr.Slider(1.0, 7.0, value=1.4, label="Petal Length (cm)"),
        gr.Slider(0.1, 2.5, value=0.2, label="Petal Width (cm)")
    ],
    outputs="markdown",
    title="ðŸŒ¸ PyCaret Iris Classifier",
    description="Adjust the sliders to predict the Iris flower species using a PyCaret model."
)

demo.launch(share=True)
```
