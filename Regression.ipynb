```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

data = fetch_california_housing(as_frame=True)
df = data.frame.copy()          # already includes target column
df.rename(columns={'MedHouseVal': 'target'}, inplace=True)

print(df.shape)
df.head()
```

```python
from pycaret.regression import *

reg = setup(
    data=df,
    target='target',
    session_id=42,
    normalize=True,
    use_gpu=True,
    fold=5
)
```

```python
best = compare_models()
tuned = tune_model(best)
final = finalize_model(tuned)
predict_model(final)
```

```python
# Quick interactive evaluation
evaluate_model(best)
```

```python
plot_model(best, plot='residuals')
plot_model(best, plot='error')
plot_model(best, plot='feature')
```

```python
tuned = tune_model(best)
final = finalize_model(tuned)
```

```python
# Predict on test data (created during setup)
preds = predict_model(final)
preds.head()

# Save model pipeline
save_model(final, 'regression_california_best')
print("âœ… Model saved as 'regression_california_best.pkl'")
```
